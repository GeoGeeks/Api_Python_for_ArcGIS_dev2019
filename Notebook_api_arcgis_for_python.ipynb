## Importar librerias
Importar las librerias que se van a usar 

```python
from arcgis import GIS

import pandas as pd
import matplotlib as plt
import numpy as np

from arcgis.features import SpatialDataFrame
```

## Conectar a cuenta arcgis online (No es obligatorio)
También se puede hacer uso de contenido público sin necesidad de hacer uso de una cuenta ArcGis

```python
gis=GIS("https://www.arcgis.com","dparra_geek","Esrico123*")
```

## Importa contenido de arcgis online con el id de ArcGis online
Se toma el contenido por medio del id 

```python
item = GIS().content.get("5479b54bd1544e3a9553f40f219da67b")
item
```

## Toma la primera capa del contenido
El contenido de Hurtos poseé dos capas, que corresponden al mapa base de la ciudad de Bogotá (capa 1) y a la ubicación de los hurtos (capa 0)

```python
flayer = item.layers[0]
```

## Convierte los datos a un dataFrame de pandas
Los datos de la capa de Hurtos se convierten a un DataFrame de Pandas y se muestra la cabeza del DataFrame

```python
sdf = pd.DataFrame.spatial.from_layer(flayer)
sdf.head()
```

## Agrupar por localidad y el total de hurtos

```python
sdf3=pd.DataFrame(sdf.groupby(['LOCALIDAD'])['TOTAL'].sum())
sdf3
```

## Datos agrupados por la caracterización del hurto

```python
pd.DataFrame(sdf.groupby(['CARACTERIZACION'])['TOTAL'].sum())
```
