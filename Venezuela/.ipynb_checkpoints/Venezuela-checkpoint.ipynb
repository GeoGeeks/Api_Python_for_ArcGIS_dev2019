```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("Libro1.csv", sep=';',dtype={'columna':'str','georreferenciada':'str'}))
df=df.dropna(axis=0, subset=['georreferenciada'])
df=df.dropna(axis=0, subset=['columna'])
df
```

```python
long=df['georreferenciada']
df['long']=0
for i in range(1734):
    df['long'][i]=(long[i])[0:3]+'.'+(long[i])[3:]
```

```python
lat=df['columna']
df['lat']=0
for i in range(1734):
    if (lat[i])[0]=='1':
        df['lat'][i]=(lat[i])[0:2]+'.'+(lat[i])[2:]
    elif (lat[i])[0]=='-':
        df['lat'][i]=(lat[i])[1:2]+'.'+(lat[i])[2:]
    elif (lat[i])[0]=='0':
        df['lat'][i]=(lat[i])[0:1]+'.'+(lat[i])[2:]
    else:
        df['lat'][i]=(lat[i])[0:1]+'.'+(lat[i])[1:]       
```

```python
df= df.drop(columns=['georreferenciada', 'Nueva','columna'], axis=1)
df
```

```python
df[['long','lat']] = df[['long','lat']].apply(pd.to_numeric)
df
```

```python
df.to_csv('venezuelafinal.csv', sep=',', encoding='utf-8')
```
